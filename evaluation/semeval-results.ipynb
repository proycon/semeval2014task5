Results
-----------------------

The code of the actual evaluation algorithm is in the Python library ``libsemeval2014task5``, supplied with the task. Below you will find the results in graphs and tables, along with the code that produced it. In the graphs, three colours show:


* *blue* - **Absolute Accuracy**: fragments predicted exactly
* *green* - **Word Accuracy**: accounting for solutions that are partially correct
* *red* - **Recall**: the number of fragments for which output was given (regardless of whether it was correct)


The word accuracy metric has been made more lenient in the latest implementation, the old version did not give any credit if only a part of the output fragment matched with only a part of the reference. The improved implementation assigns a score according to the longest *consecutive* matching substring between output fragment and reference fragment. Word accuracy is computed as:

$$wac = \frac{|longestsubmatch(output,reference)|}{max(|output|,|reference|)}$$

Note that this always has a positive effect on all word-accuracy scores, compared to the earlier version. It can never produce a lower score than the older script. 

Scores are also matched against all alternatives in the reference output, the best score counts. Manual care was taken to enrich the gold standard with valid alternatives from participant systems that were not considered yet. 

For out-of-five evaluation, only the best score from the five options counts, again matched against all reference alternatives.

Furthermore, the evaluation script has been adapted to be as lenient as possible with tokenisation differences. Special care was given to contractions, some rewrite rules were added to rewrite ``de el`` and ``a el`` in Spanish participant output back to ``del`` and ``al``. Moreover, any trailing or leading punctuation was stripped from the fragments prior to evaluation. Evaluation is case-sensitive.



```python
#Code to generate the evaluation output seen here

import libsemeval2014task5
import matplotlib.pyplot as plt
import glob
import os
import numpy as np
from IPython.display import HTML

DIR = "/home/proycon/semeval2014task5-pub/evaluation/"
os.chdir(DIR)

def evaluate():
    for f in sorted(glob.glob('*xml')):
        if f.find('matrex') != -1:
            #skip intermediate files
            continue
        fields = os.path.basename(f).split('.')
        team = fields[0]
        L1, L2 = fields[1].split('-')
        run = fields[2]
        
        #Find evaluation type, can be best or oof
        #If not specified, the file will be evaluated according to both metrics
        if fields[3] == 'xml':
            eval = 'both'
        else:
            eval = fields[3]
    
        if eval == 'both' or eval == 'best':
            #BEST Evaluation
            print("Processing " + f  + " (BEST)")
            if eval == 'both':
                evalf = team + '.' + L1 + '-' + L2 + '.' + run + '.best.xml'
                os.rename(f, evalf) #temporary rename so output files will be consistently named
            else:
                evalf = f
            cmd = "semeval2014task5-evaluate -I --ref ../corpus/" + \
            L1 + "-" + L2 + ".gold.tokenised.xml --out " + evalf + ">/dev/null 2>/dev/null"
            r = os.system(cmd)
            if eval == 'both':
                os.rename(evalf, f) #undo temporary rename
            if r != 0:
                print("Failed prematurely!")
                return
        if eval == 'both' or eval == 'oof':
            #OOF Evaluation
            print("Processing " + f  + " (OOF)")
            if eval == 'both':
                evalf = team + '.' + L1 + '-' + L2 + '.' + run + '.oof.xml'
                os.rename(f, evalf)  #temporary rename so output files will be consistently named
            else:
                evalf = f
            cmd = "semeval2014task5-evaluate -I -a --ref ../corpus/" + \
            L1 + "-" + L2 + ".gold.tokenised.xml --out " + evalf + ">/dev/null 2>/dev/null"
            r = os.system(cmd)
            if eval == 'both':
                os.rename(evalf, f) #undo temporary rename
            if r != 0:
                print("Failed prematurely!")
                return
    print("Done")


evaluate()

data = {}

%matplotlib inline
```

```python
#Read score summaries into memory

teams = set()
for filename in glob.glob("*.summary.score"):
    with open(filename) as f:
        team, langs, run, evtype, _,_ = os.path.basename(filename).split('.',5)
        if not langs in data:
            data[langs] = {}
        if not evtype in data[langs]:
            data[langs][evtype] = {}
        if not team + '-' + run in data[langs][evtype]:
            data[langs][evtype][team + '-' + run] = {}
        d = data[langs][evtype][team + '-' + run]
        teams.add(team)
        f.readline()
        d['ac'], d['wac'],d['rec'] = ( float(x) for x in f.readline().split(" ") )
```

```python
#Output graphs

def graph(langs, evtype):
    values = []
    teams = sorted(data[langs][evtype].keys())
    for team in teams:
        values.append(data[langs][evtype][team])
        
    index = np.arange(len(values))
    bar_width = 0.25
    
    plt.figure(figsize=(19,8))
    plt.bar(index, [x['ac'] for x in  values] , bar_width,color='#6071B3',label="Accuracy")
    plt.bar(index + bar_width, [x['wac'] for x in values], bar_width,color='#60B362',label="Word Accuracy")
    plt.bar(index + 2*bar_width, [x['rec'] for x in values], bar_width,color='#B36060',label="Recall")
    plt.xticks(index + bar_width, [x.replace("-","\n") for x in  teams])
    plt.xlabel("Systems")
    plt.ylabel("Score")
    plt.grid()
    plt.title(langs + " (" + evtype + ")")
    #plt.legend()
    plt.show()    

graph('en-es','best')
graph('en-es','oof')
graph('en-de','best')
graph('en-de','oof')
graph('fr-en','best')
graph('fr-en','oof')
graph('nl-en','best')
graph('nl-en','oof')
```

```python
#Output Table

s = ""
for langs in sorted(data):
    for evtype in sorted(data[langs]):
        s += "<h2>" + langs + " (" + evtype + ")</h2>"
        s += "<table>"
        s += "<tr><th></th><th>Accuracy</th><th>Word Accuracy</th><th>Recall</th></tr>"
        for team, d in sorted(data[langs][evtype].items(), key=lambda x: -1 * x[1]['wac'] ):
            s += "<tr><th>" + team + "</th><td>" + str(round(d['ac'],3)) + "</td><td>" + str(round(d['wac'],3)) + "</td><td>" + str(d['rec']) + "</tr>"
        s += "</table>"

HTML(s)
```

```python
#CSV output
s = ""
f = open(DIR + '/semeval2014task5-results.csv','w',encoding='utf-8')
f.write("Language Pair\tEvaluation Type\tTeam\tAccuracy\tWord Accuracy\tRecall\n")
for langs in sorted(data):
    for evtype in sorted(data[langs]):
        for team, d in sorted(data[langs][evtype].items(), key=lambda x: -1 * x[1]['wac'] ):
            s = langs + "\t" + evtype + "\t" + team + "\t" + str(round(d['ac'],3)) + "\t" + str(round(d['wac'],3)) + "\t" + str(d['rec'])+"\n"
            f.write(s)
f.close()

```
